```python
import pandas as pd

data={
    "product":["wireless Earbuds","smartphone","laptop","smartwatch","Bluetooth Speaker"],
    "Brand":["soundmax","technova","bytecore","timetrack","echoboom"],
    "price":[2999,15999,52999,49999,1999],
    "stock":[50,30,40,20,60],
    "BestSeller":[True,False,False,True,True]
}

df=pd.DataFrame(data)
print(df)
```

```python
print("Columns in dataframe:",df.columns)
```

```python
print("Dataframe Info:")
print(df.info())
```

```python
print("Head:\n",df.head(3))
```

```python
print("Tail:\n",df.tail(2))
```

```python
print(df)
print(df.loc[1,"product"])
print(df.loc[4,"product"])
print(df.loc[4,"stock"])

print(df.iloc[4,0])
print(df.iloc[2,1])
print(df.iloc[3,3])
```

```python
print("using iloc(select 4th row):\n",df.iloc[3])
```

```python
print(df)
print("*********************using loc********************\n",df.loc[df['BestSeller']==True])
print("*********************using loc********************\n",df.loc[df['stock']<40])
print("*********************using loc********************\n",df.loc[df['price']<10000])
```

```python
df_dropped=df.drop(columns=["stock"])
print("After Dropping 'stock' Column:\n",df_dropped)
```

```python
df_renamed=df.rename(columns={"price":"cost"})
print("After Renaming 'price' to 'cost':\n",df_renamed)
```

```python
df_grouped=df.groupby("Brand").agg({"price":"sum"})
print(df_grouped)
```

```python
df_grouped=df.groupby("Brand").agg({"price":["mean","max","min"]})
print(df_grouped)
```

```python
import pandas as pd
import numpy as np

data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [np.nan, 2, 3, np.nan, 5],
    'C': [10, np.nan, np.nan, 40, 50]
}

df = pd.DataFrame(data)
print("Original DataFrame with Missing Values:")
print(df)

```

```python
df_interpolate = df.interpolate()
print("\nDataFrame after filling NaN values using interpolation:")
print(df_interpolate)
```

```python
df_dropna = df.dropna()
print("\nDataFrame after dropping rows with NaN values:")
print(df_dropna)
```

```python
df_fillna = df.fillna(0)
print("\nDataFrame after filling NaN values with 0:")
print(df_fillna)
```
