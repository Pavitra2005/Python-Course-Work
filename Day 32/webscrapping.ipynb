```python
import  pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
URL="http://timely-sunshine-e821b3.netlify.app/"
```

```python
page=requests.get(URL)
```

```python
page.status_code
```

```python
htmlCode = page.text
print(htmlCode)
```

```python
soup=BeautifulSoup(htmlCode,'html.parser')
print(soup)
```

```python
content=soup.find_all('div',attrs={'class':'a'})
for i in content:
  print("Div:",i,end='\n\n\n\n\n')
```

```python
content=soup.find('div',attrs={'class':'name'})
text=content.text.strip()
print(text)
```

```python
content=soup.find('div',attrs={'class':'price'})
text=content.text.strip()
print(text)
```

```python
content=soup.find('div',attrs={'class':'image'})
print(content)
```

```python
img_tag=content.find('img')
print(img_tag)
```

```python
items=soup.find_all('div',class_='a')
```

```python
img_src=img_tag.get('src')
print(img_src)
```

```python
items=soup.find_all('div',class_='a')

names=[]
prices=[]
images=[]

for item in items:
  name=item.find('div',class_='name').text.strip()
  price=item.find('div',class_='price').text.strip()
  image_tag=item.find('div',class_='image').find('img')
  image_src=image_tag['src']if image_tag else None

  names.append(name)
  prices.append(price)
  images.append(image_src)

  print("Names:",names)
  print("Prices:",prices)
  print("Images:",images)
```

```python
df=pd.DataFrame({'Product_name': names,'MRP': prices,'Image_SRC': images})
```
