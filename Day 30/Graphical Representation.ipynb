```python
import matplotlib.pyplot as plt

hours=[1,2,3,4,5,6,7,8]
visitors=[50,60,70,90,110,130,160,180]

plt.plot(hours,visitors,
         color="blue",
         marker="s",
         linestyle="-",
         linewidth=2,
         alpha=1,
         label="Visitors"
         )
plt.title("Website Visitors per hour")
plt.xlabel("Hour of day")
plt.ylabel("Number of visitors")
plt.legend()
plt.grid(True,linestyle="-",alpha=1)
plt.show()
```

```python
import matplotlib.pyplot as plt

brands=["Apple","samsung","oneplus","oppo","realme"]
market_share=[30,25,20,10,15]

plt.pie(market_share,
        labels=brands,
        autopct="%d%%",
        startangle=270,
        colors=["gold","skyblue","lightcoral","lightgreen","violet"],
        explode=[0,0,0,0.3,0],
        shadow=True)

plt.title("Mobile Market Share 2025")
plt.show()
```

```python
import matplotlib.pyplot as plt

scores=[
    [70,80,95,92,93,99],
    [58,45,32,54,43,56],
    [90,91,89,67,56,45],
    [89,67,78,98,65,64],
    [78,90,93,98,57,54]
]

plt.boxplot(scores,
            patch_artist=False,
            notch=False,
            vert=True,
            labels=["A","B","C","D","E"])
plt.title("Exam scores distribution Across Classes")
plt.xlabel("Classes")
plt.ylabel("Scores")
plt.grid(True,linestyle="-",alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt

hours=[1,2,3,4,5,6,7,8]
visitors=[50,60,70,90,110,130,160,180]

plt.plot(hours,visitors,
         color="blue",
         marker="s",
         linestyle="-",
         linewidth=2,
         alpha=1,
         label="Visitors"
         )
plt.fill_between(hours,visitors,
                 color="skyblue",
                 alpha=1,
                 label="Website Visitors")

plt.title("Website Visitors per hour")
plt.xlabel("Hour of day")
plt.ylabel("Number of visitors")
plt.legend()
plt.grid(True,linestyle="-",alpha=1)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

data=np.array([
    [30,32,35,36,29,27,34],
    [24,23,34,35,34,36,45],
    [23,45,56,65,54,43,33],
    [27,45,46,51,23,33,34],
    [34,43,42,21,23,33,35]
])

days=["mon","tue","wed","thu","fri","sat","sun"]
weeks=["Week1","Week2","Week3","Week4","Week5"]

sns.heatmap(data,
            annot=True,
            fmt=".2f",
            cmap="YlOrRd",
            linewidth=0.5,
            xticklabels=days,
            yticklabels=weeks)

plt.title("Weekly Temperature heatmap",fontsize=14,fontweight="bold")
plt.xlabel("Days of the week")
plt.ylabel("Weeks")
plt.show()
```

```python
import matplotlib.pyplot as plt

ad_spend = [2, 4, 5, 7, 8, 10, 11, 12]
sales = [27, 25, 28, 35, 40, 48, 50, 52]
profit = [200, 250, 90, 350, 40, 80, 500, 520]
colors = [8,1,2,3,5,6,7,4]

plt.scatter(ad_spend, sales,
            s=profit,
            c=colors,
            cmap='plasma',
            alpha=0.8,
            edgecolors='black',
            linewidths=1.2
           )

plt.title("Advertising Spend vs Sales (with Color & Size)", fontsize=14, fontweight="bold")
plt.xlabel("Ad Spend ($1000s)", fontsize=12)
plt.ylabel("Sales ($1000s)", fontsize=12)

plt.colorbar(label="Sales Value")

plt.grid(True, linestyle="--", alpha=0.2)
plt.show()
```

```python
import matplotlib.pyplot as plt

days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
sales = [120, 135, 150, 160, 170, 200, 220]
profit = [30, 40, 45, 50, 55, 70, 80]

plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.plot(days, sales, marker="o", color="blue")
plt.title("Sales Over Days")
plt.grid(True)

plt.subplot(2, 2, 2)
plt.bar(days, sales, color="orange")
plt.title("Sales Bar Chart")

plt.subplot(2, 2, 3)
plt.scatter(sales, profit, color="green", s=100, edgecolors="black")
plt.title("Sales vs Profit")

plt.subplot(2, 2, 4)
plt.pie(sales, labels=days, autopct="%1.1f%%", startangle=90)
plt.title("Sales Distribution")

plt.tight_layout()
plt.show()
```
